```
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import style
```

```
lending_club_data = pd.read_csv('https://raw.githubusercontent.com/cjflanagan/sample_data/master/loan.csv')
lc_data = lending_club_data
```

```
lc_data.head(5)
#lc_data.shape  # 10966 rows, 145 columns
```

```
f, axes = plt.subplots(2,4, figsize=(20,6), squeeze=True)  #, sharey=True)  #, subplot_kw=dict(polar=True))  ax.ticklabel_format(style='sci', scilimits=(0,0), axis='both')
sns.distplot(lc_data.loan_amnt, rug=True, kde=True,  ax=axes[0,0])
sns.distplot(lc_data.funded_amnt, rug=True, kde=True,  ax=axes[0,1])
sns.distplot(lc_data.int_rate, rug=True, kde=False, ax=axes[0,2])
sns.distplot(lc_data.installment, rug=True, kde=True, ax=axes[0,3])
sns.distplot(lc_data.loan_amnt, rug=True, kde=True,  ax=axes[1,0])
sns.distplot(lc_data.funded_amnt, rug=True, kde=True,  ax=axes[1,1])
sns.distplot(lc_data.int_rate, rug=True, kde=True, ax=axes[1,2])
sns.distplot(lc_data.installment, rug=True, kde=True, ax=axes[1,3])
plt.tight_layout()
```

```
with sns.axes_style("white"):
    sns.jointplot(data=lc_data, x='loan_amnt',y='installment', kind="hex", color="r")
```

```
style.available  # style is a global variable  BE CAREFUL!
#style.use('grayscale')
style.use('fivethirtyeight')
```

```
# Colorblind-friendly colors
colors = [[0,0,0], [230/255,159/255,0], [86/255,180/255,233/255], [0,158/255,115/255],
          [213/255,94/255,0], [0,114/255,178/255]]
          
        
#edu_plot = lc_data.plot(x='loan_amnt',y='int_rate', figsize=(15, 10), legend=False, color='green', marker='o', linestyle='None')
edu_plot = lc_data.plot(x='loan_amnt',y='int_rate', figsize=(15, 10), legend=False, color=(0,114/255,178/255), marker='o', linestyle='None')

#edu_plot = sns.scatterplot(data=lc_data, x='loan_amnt',y='int_rate', figsize=(15, 10), legend=False)
#edu_plot = lc_data.scatter(x='loan_amnt',y='int_rate', figsize=(15, 10), legend=False)

edu_plot.axhline(y=0, color='black', alpha=0.7, linewidth=1.5)
edu_plot.tick_params(axis='both', which='major', labelsize=18)
edu_plot.set_yticklabels(labels = ['santa', '0 ', '5 ', '10 ', '15 ', '20 ', '25 ', '30'])
edu_plot.set_xlabel('Loan Amount ($)')
edu_plot.set_ylabel('Interest Rate (%)')


edu_plot.set_xlim(left=0, right=40000)

edu_plot.text(x=0,
              y=-5.5,
              s="""© CS06 Data Viz with Python             """
                + """                                                             """
                + """                                 Source: Kaggle""",
              fontsize=17,
              color='#f0f0f0',
              backgroundcolor='grey'
              )


# Adding a title and a subtitle
edu_plot.text(x = 0, y = 35, s = "538 of Lending Club",
              fontsize = 26, weight = 'bold', alpha = .75)
edu_plot.text(x = 0, y = 33,
              s = 'Interest Rate on Loans',
              fontsize = 19, alpha = .85)


# Add colored labels
#edu_plot.text(x = 30000, y = 43, s = 'Agriculture', color = colors[0], weight = 'bold', rotation = 33,
#              backgroundcolor = '#f0f0f0')



plt.show()
```
