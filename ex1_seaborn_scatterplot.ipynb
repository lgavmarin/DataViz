```
import pandas as pd
import numpy as np

from vega_datasets import data
movies_data = data.movies()
movies_data.head()
```

```
#pd.DatetimeIndex(movies_data.Release_Date).year
#pd.DatetimeIndex(movies_data.Release_Date).day
movies_data['year']=pd.DatetimeIndex(movies_data.Release_Date).year
movies_data.head()
```

```
movies_data[movies_data.year > 2019]
```

```
import matplotlib.pyplot as plt
import seaborn as sns

#plt.figure(figsize=(10,8))
plt.figure()
#sns.scatterplot(x='IMDB_Rating', y='IMDB_Votes', size='Production_Budget', hue='US_Gross', alpha=0.5, data=movies_data)
#sns.scatterplot(x='IMDB_Rating', y='IMDB_Votes', hue='US_Gross', alpha=0.5, data=movies_data)
sns.scatterplot(x='US_Gross', y='IMDB_Rating', alpha=0.5, data=movies_data)
plt.title('Movies')
plt.show

plt.figure()
sns.scatterplot(x='IMDB_Rating', y='IMDB_Votes', alpha=0.5, data=movies_data)
plt.title('Movies')
plt.show

plt.figure()
sns.scatterplot(y='US_Gross', x='Production_Budget', hue='IMDB_Rating', alpha=0.3, data=movies_data)
plt.title('Movies')
plt.show

plt.figure()
sns.scatterplot(x='US_Gross', y='IMDB_Rating', size='Production_Budget', alpha=0.5, data=movies_data)
plt.title('Movies')
plt.show

plt.figure()
sns.scatterplot(x='US_Gross', y='Worldwide_Gross', hue='IMDB_Rating', alpha=0.5, data=movies_data)
plt.title('Movies')
plt.show

plt.figure()
sns.scatterplot(x='Release_Date', y='Production_Budget', alpha=0.5, data=movies_data)
plt.title('Movies')
plt.show
```

```
movies_data['Worldwide_Gross_Log']=np.log(movies_data.Worldwide_Gross)
```
